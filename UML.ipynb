## Breaking RSA with Shor's Algorithm

```python
import iplantuml
```

```python
%%plantuml

@startuml
actor Alice as a
actor Eve as e
actor Bob as b

note over of b : **Has** a confidential message: M
a -> a : Alice select: p & q
a -> a : Alice calculate: N = p * q
a -> a : Alice calculate: ϕ(N) = (p-1)*(q-1)
a -> a : Alice calculate: e → gcd(ϕ(N),e)=1; 1<e<ϕ(N)
a -> a : Alice calculate: d ≡ e^-1 (mod ϕ(N))
a -> b : Alice send public key to Bob: (N,e)
a -> e : Eve get public key: (N,e)
b -> b : Bob encrypt the message: C ≡ M^e (mod N)
b -> a : Bob send the ciphertext Alice: C
b -> e : Eve intercept the ciphertext: C
a -> a : Alice decrypt the ciphertext: M ≡ C^d (mod N)
note over of a : **Obtain** the confidential message: M
e -> e : Eve use Shor's Algorithm: factor N into p & q
e -> e : Eve claculate: ϕ(N) = (p-1)*(q-1)
e -> e : Eve calculate: d ≡ e^-1 (mod ϕ(N))
e -> e : Eve decrypt the ciphertext: M ≡ C^d (mod N)
note over of e: **Obtain** the confidential message: M

@enduml
```
